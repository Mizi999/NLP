```python
#yaz 21.03 - den 20.06
#yay 21.06 - den 20.09
#payiz 21.09-den 20.12
#qis 21.12-den 20.03
def get_season(date):
  mounth = date.month
  day = date.day
  if (mounth == 3 and day >=21) or (mounth == 6 and day <=20):
    return 'ilkbahar'
  elif (mounth == 6 and day >=21) or (mounth == 9 and day <=20):
    return 'yaz'
  elif (mounth == 9 and day >=21) or (mounth == 12 and day <=20):
    return 'sonbahar'
  else:
    return 'kış'
```

```python
df_2022['season'] = df_2022['date'].apply(lambda x: get_season(x))
```

```python
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_2022['day_of_week'] = df_2022['date'].apply(lambda x: days_of_week[x.weekday()])
```

```python
def get_period(hour):
  hour1 = hour.hour
  if hour in [22,23,0,1,2]:
    if hour1 == 2 and hour.minute != 0:
      return '02:00-06:00'
    else:
      return '22:00-02:00'
  elif hour1 in [3,4,5,6]:
    if hour1 == 6 and hour.minute != 0:
      return '06:00-10:00'
    return '03:00-06:00'
  elif hour1 in [7,8,9,10]:
    if hour1 == 10 and hour.minute !=0:
      return '10:00-14:00'
    return '06:00-10:00'
  elif hour1 in [11,12,13,14]:
    if hour1 == 14 and hour.minute !=0:
      return '14:00-18:00'
    return '10:00-14:00'
  elif hour1 in [15,16,17,18]:
    if hour1 == 18 and hour.minute !=0:
      return '18:00-22:00'
    return '14:00-18:00'
  else:
    return '22:00-02:00'
```

```python
df_2022['hours_of_period'] = df_2022['date'].apply(lambda x: get_period(x))
```

```python
def vizuallation_and_preprocessing(text,plot = True):
    text = text.str.lower()
    text = text.str.replace('[^\w\s]','',regex=True)
    text = text.str.replace('\d+','',regex=True)
    text = text.str.replace('\n','',regex=True)
##########################################
    stop_words = nltk.corpus.stopwords.words('turkish')
    text = text.apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))
    temp_df = pd.Series(' '.join(df_2022['tweet']).split()).value_counts().sort_values(ascending=False)
    drop = temp_df[temp_df<7]
    text = text.apply(lambda x: ' '.join(word for word in x.split() if word not in drop))
##########################################
    text1 = text.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
    if plot:
      number_of_words = pd.Series(' '.join(text1).split()).value_counts().reset_index()
      number_of_words.columns = ['word','count']
      number_of_words=number_of_words.sort_values(by='count',ascending=False)
      plt.subplot(1,2,1)
      sns.barplot(x=number_of_words['word'].head(10),y=number_of_words['count'].head(10))
      plt.xlabel('Word')
      plt.ylabel('Count')
      plt.xticks(rotation=90)
      plt.title('Top 10 Words')
      plt.subplot(1,2,2)
      text = ' '.join(text1)
      wordcloud = WordCloud(width=800,height=500,background_color='white').generate(text)
      plt.imshow(wordcloud)
      plt.axis('off')
      plt.title('Word Cloud')
      plt.tight_layout()
      plt.show()
    return text1
```

```python
df_2022['tweet'] = vizuallation_and_preprocessing(df_2022['tweet'],plot=False)
```

```python
mean_accuracy = []
for i in [log,dt,rf]:
  score = cross_val_score(i,X,y,cv=5)
  mean_accuracy.append(score.mean())
```
